```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
grade = pd.read_csv('board2.csv')
grade
```

```python
Instru = grade.loc[(grade['Track'] == 'Instrumentation')
          & (grade['Hometown'] == 'Luzon')
          & (grade['Electronics'] > 70), ('Name','Electronics','GEAS')].reset_index(drop=True)
Instru.to_csv('Instru.csv', index = False)

Instru
```

```python
grade['Average'] = grade[['Math','Electronics','GEAS','Communication']].mean(axis=1)
grade
```

```python
Mindy = grade.loc[(grade['Hometown'] == 'Mindanao')
           & (grade['Gender'] == 'Female')
           & (grade['Average'] >= 55),['Name', 'Track', 'Electronics', 'Average']].reset_index(drop=True)

Mindy
```

```python
ave_Track = grade.groupby('Track')['Average'].mean()
ave_Gender = grade.groupby('Gender')['Average'].mean()
ave_Hometown = grade.groupby('Hometown')['Average'].mean()
```

```python
ave_Track.plot(kind='bar', color='red', title='Average Grade by Track', xlabel='Track', ylabel='Average Grade')
plt.show()
```

```python
ave_Gender.plot(kind='bar', color='purple', title='Average Grade by Gender', xlabel='Gender', ylabel='Average Grade')
plt.show()
```

```python
ave_Hometown.plot(kind='bar', 
                  color='blue', 
                  title='Average Grade by Hometown', 
                  xlabel='Hometown', 
                  ylabel='Average Grade')
plt.show()
```
